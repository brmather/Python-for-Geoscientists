# Making multi-pane maps

Create a figure and a set of subplots.

This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import numpy as np
import cartopy.crs as ccrs
import gplately

%matplotlib inline
```

```python

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 9), subplot_kw={'projection':ccrs.PlateCarree()})
```

## GridSpec

For more refined control

```python
gs = GridSpec(nrows=2, ncols=2)
base_proj = ccrs.PlateCarree()


fig = plt.figure(figsize=(8,9))

ax1 = fig.add_subplot(gs[0,0], projection=base_proj)
ax2 = fig.add_subplot(gs[0,1], projection=base_proj)
ax3 = fig.add_subplot(gs[1,0], projection=base_proj)
ax4 = fig.add_subplot(gs[1,1], projection=base_proj)
```

## Adding new axes

```python
fig = plt.figure(figsize=(8,9))

ax1 = fig.add_subplot(111, projection=base_proj)

ax2 = fig.add_axes([0,0,0.5,0.01], projection=base_proj)
```
